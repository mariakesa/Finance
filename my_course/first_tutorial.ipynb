```python
import numpy as np
import matplotlib.pyplot as plt

# number of samples
N = 1_000_000

# ---- Step 1: sample X ----
# f_X(x) = x/800 on (0,40)
# This is a triangular distribution
X = 40 * np.sqrt(np.random.rand(N))

# ---- Step 2: sample Y|X ----
Y = np.random.rand(N) * (3 * X)

# ---- Step 3: compute Z ----
Z = Y - X

# ---- theoretical density ----
z1 = np.linspace(-40, 0, 500)
fz1 = (40 + z1) / 2400

z2 = np.linspace(0, 80, 500)
fz2 = (40 - z2/2) / 2400

# ---- plot ----
plt.figure(figsize=(8,5))

plt.hist(Z, bins=300, density=True, alpha=0.5, label="simulation")

plt.plot(z1, fz1, linewidth=3, label="theory")
plt.plot(z2, fz2, linewidth=3)

plt.xlabel("z")
plt.ylabel("density")
plt.title("Distribution of Z = Y - X")
plt.legend()
plt.show()
```
