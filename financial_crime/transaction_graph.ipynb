```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat=pd.read_csv('transactions.csv')
```

```python
def make_adjacency_list(dat, multi=False):
    source_accounts=dat['src_account']
    destination_accounts=dat['dst_account']
    adjacency_map={}
    for src in set(source_accounts):
        src_indices=source_accounts[source_accounts==src].index
        dests=destination_accounts[src_indices]
        if multi==True:
            adjacency_map[src]=list(sorted(dests))
        else:
            adjacency_map[src]=list(set(dests))
    return adjacency_map

adj_list=make_adjacency_list(dat)
print(adj_list)
def bfs(adj_list,root='acct_7d2bc231cfd5'):
    #Queue-first in first out, list pop is first in last out. pop(0) pops first.
    explored={}
    queue=[root]
    explored[root]=True
    while queue:
        node=queue.pop(0)
        neighbors=adj_list.get(node, [])
        for n in neighbors:
            if n not in explored:
                explored[n]=True
                queue.append(n)           
    print(explored.keys())
bfs(adj_list)
```
