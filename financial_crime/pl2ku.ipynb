```python
import numpy as np
import matplotlib.pyplot as plt

# simulate normal behavior
n = 100000

X = np.random.lognormal(mean=7, sigma=1, size=n)   # amounts
T = np.random.exponential(scale=2, size=n)         # hours since last txn

Z = X / T  # derived variable

plt.hist(Z[:100], bins=200, density=True)
plt.xlim(0,5000)
plt.title("Derived distribution: money velocity")
plt.show()

# suspicious example
z_obs = 10000

p = np.mean(Z > z_obs)

print("Probability of observing this velocity:", p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# simulate realistic transaction amounts
transactions = np.random.lognormal(mean=7, sigma=1, size=100000)

# extract first digits
first_digits = [int(str(int(x))[0]) for x in transactions]

counts = np.bincount(first_digits)[1:]

plt.bar(range(1,10), counts / counts.sum())
plt.title("First Digit Distribution")
plt.xlabel("Digit")
plt.ylabel("Frequency")
plt.show()
```
