```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat=pd.read_csv('transactions.csv')
```

```python
import numpy as np

def make_risk_weighted_graph(
    dat,
    amount_col="amount",
    off_hours_col="is_off_hours",
    cross_border_col="is_cross_border",
    alpha=1.0,
    beta=2.0,
    gamma=2.0
):
    
    adjacency = {}

    for _, row in dat.iterrows():

        src = row["src_account"]
        dst = row["dst_account"]

        amount = row[amount_col]
        off_hours = row[off_hours_col]
        cross_border = row[cross_border_col]

        risk = (
            alpha * np.log(amount + 1) +
            beta * off_hours +
            gamma * cross_border
        )

        adjacency.setdefault(src, []).append((dst, risk))

    
    return adjacency

adj_graph=make_risk_weighted_graph(dat)

adj_graph
```
